```python
# x = x/255
# print(x)
def shuffle(a, b):
    
    sf = len(x)
    
    for i in range(sf):
        
        j = np.random.randint(0, sf-1)
        
        w = a[i]
        w1 = b[i]
        
        a[i] = a[j]
        b[i] = b[j]
        
        a[j] = w
        b[j] = w1


    return x, y
x, y = shuffle(x, y)

print(x.shape)
print(y.shape)
    
# print(x)
```

```python
x = x/255.0
```

```python

image_shape = (16, 50, 50, 1)
num_classes = 52

model = Sequential()


conv1 = Conv3D(48, kernel_size=(3, 3, 3),
                 activation='relu',
                 input_shape=image_shape)

model.add(conv1)

pool_layer1 = MaxPooling3D(pool_size=(3,3,3))


model.add(pool_layer1)


conv2 = Conv3D(256,(3,3,3), activation='relu', padding='same')

model.add(conv2)

pool_layer2 = MaxPooling3D(pool_size=(3,3,3))

model.add(pool_layer2)


conv3 = Conv3D(512,(1,3,3), activation='relu')

model.add(conv3)

conv4 = Conv3D(512,(1,3,3), activation='relu')

model.add(conv4)

# model.add(Conv3D(512,(1,1,1), activation='relu'))

# model.add(Conv3D(256,(1,3,3), activation='relu'))

model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(num_classes, activation='softmax'))


model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(lr=0.0000001),
              metrics=['accuracy'])

model.summary()
```

```python

batch_size = 16
epochs = 10
model.fit(x, y,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_split=0.2, shuffle=True)
```
